```python
import gradio as gr
import tensorflow as tf 
import numpy as np
import requests
import cv2
from tensorflow.keras.models import load_model
```

Bianry Calssification

```python
model = load_model("/content/drive/MyDrive/Project/Binary Classification/Magnification Independent/Sigmoid/MobileNet/mobilenet_sigmoid.h5")
```

```python
model.summary()
```

```python
labels = ['Benign','Malignant']
```

```python
def model1(image):
  a=[]
  image = image.reshape((-1, 128, 128, 3)) 
  a.append(np.array(image))
  image = a  
  prediction = model.predict(image).flatten()
  return {labels[i]: float(prediction[i]) for i in range(2)}
```

```python
image = gr.inputs.Image(shape = (128, 128)) 
label = gr.outputs.Label(num_top_classes = 2)
```

```python
gr.Interface(model1, inputs = image ,outputs = label ,title = "Magnification Independent Binary classification using MobileNetV2 and Sigmoid as an activation function",capture_session = True).launch(share=True)
```
